```python
import pandas as pd

print(pd.__version__)
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.groupby('species')['petal_length'].count().to_frame()
```

```python
df.groupby('species')['petal_length'].mean().to_frame()
```

```python
df.groupby('species')['petal_length'].median().to_frame()
```

```python
df.groupby('species')['petal_length'].agg(['count', 'mean', 'median'])
```

```python
df.groupby('species')['petal_length'].describe()
```
